```python
import pandas as pd
# Config
df_train_df_path = "/Users/david/Nextcloud/Dokumente/Education/Uni Bern/Master Thesis/Analyzing Financial Climate Disclosures with NLP/Labelling/annual reports/Firm_AnnualReport_Training_DF.csv"
df_train_ts_path = "/Users/david/Nextcloud/Dokumente/Education/Uni Bern/Master Thesis/Analyzing Financial Climate Disclosures with NLP/Labelling/annual reports/Firm_AnnualReport_Training_TS.csv"
df_test_df_path = "/Users/david/Nextcloud/Dokumente/Education/Uni Bern/Master Thesis/Analyzing Financial Climate Disclosures with NLP/Labelling/annual reports/Firm_AnnualReport_Test_DF.csv"
df_test_ts_path = "/Users/david/Nextcloud/Dokumente/Education/Uni Bern/Master Thesis/Analyzing Financial Climate Disclosures with NLP/Labelling/annual reports/Firm_AnnualReport_Test_TS.csv"
```

```python
df_train_df = pd.read_csv(df_train_df_path, index_col=0)
df_train_ts = pd.read_csv(df_train_ts_path, index_col=0)
df_test_df = pd.read_csv(df_test_df_path, index_col=0)
df_test_ts = pd.read_csv(df_test_ts_path, index_col=0)

df_train = df_train_df.merge(df_train_ts,left_index=True, right_index=True)
df_test = df_test_ts.merge(df_test_ts,left_index=True, right_index=True)
df_train["should_label"] = df_train[["should_label_x","should_label_y"]].any(axis=1)
df_train["is_labelled"] = df_train[["is_labelled_x", "is_labelled_y"]].any(axis=1)
df_test["should_label"] = df_test[["should_label_x","should_label_y"]].any(axis=1)
df_test["is_labelled"] = df_test[["is_labelled_x", "is_labelled_y"]].any(axis=1)
```

```python
print("Median train: ", df_train.query("is_labelled == True").year_x.median())
print("Median test: ", df_test.query("is_labelled == True").year_x.median())
```
