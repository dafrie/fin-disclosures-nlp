## SQuAD 2.0 Benchmark scores over time

```python
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setup seaborn
sns.set_theme(style="ticks", rc={'text.usetex' : True})
sns.set_context("paper")
df = pd.read_csv("/Users/david/Nextcloud/Dokumente/Education/Uni Bern/Master Thesis/Analyzing Financial Climate Disclosures with NLP/Data/SQuAD/squadv2_explorer_data.csv", parse_dates=['Date'])

human_score = df.loc[0]
model_scores = df.loc[1:]

ax = sns.scatterplot(data=model_scores, x="Date", y="F1", zorder=2)
ax.axhline(human_score.F1, c='grey', linestyle='dashed', label="Human Performance", zorder=1)
ax.set(xticks=[datetime(2018, 7, 1), datetime(2019,1,1), datetime(2019, 7, 1), datetime(2020,1,1), datetime(2020,7,1)], xlabel='Date of submission', ylabel='F1-Score', title="F1-Scores submitted to the SQuAD 2.0 Leaderboard over time")
ax.set_xlim(datetime(2018, 1, 1), datetime(2020, 12, 31))
ax.legend()
fig = ax.get_figure()
fig.savefig('/Users/david/Nextcloud/Dokumente/Education/Uni Bern/Master Thesis/Analyzing Financial Climate Disclosures with NLP/Thesis/figures/squadv2_evolution.pdf', format='pdf', bbox_inches='tight')
```
